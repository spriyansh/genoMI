### 1. ```plot()```
* ```plot()``` function is used for making basic types of plot
* It accepts the "x" and "y" coordinates to produce 2D plot
* Attributes such as labels and legends can be added to the plot using various methods

````python
# Values 
x = np.arange(1, 51, 1)
y = np.arange(1, 101, 2)

# Plotting 
plt.plot(x,y, "r--") # r--: red colour and dotted line
plt.xlim(1,100)
plt.ylim(1,200)
plt.title("Line Plot")
plt.xlabel("X-Label"), plt.ylabel("Y-Label")
````

### 2. ```imshow()```
* It accepts matrix and creates a heatmap representation
* ```cmap = ''```, argument can be used to choose between various colour schemes
* ```colorbar()```, generates color bar for the heatmap

````python
# Making Matrix
m = np.random.randint(1,101,(10,10))
m

# Heatmap
plt.imshow(m, cmap = 'gist_rainbow')
plt.colorbar()
````

### 3. Plotting directly from the dataFrame
* ```plot()```, directly used with the pandas dataFrame generates plots

````python
# Reading csv
df = pd.read_csv("df_without_index.csv")

# Plotting
df.plot()
````
